NOMOR 1 DENGAN GAUSS SEIDELL

```python
import numpy as np

print("=== System 1: Gauss-Seidel Method ===")
A1_iter = np.array([[2, 3, -1],
                    [1, 1, 2],
                    [3, 2, 1]], dtype=float)


diag = np.diag(np.abs(A1_iter))
off_diag = np.sum(np.abs(A1_iter), axis=1) - diag

if np.all(diag > off_diag):
    print("Matrix is diagonally dominant")
else:
    print("NOT diagonally dominant")


x1, y1, z1 = 5.0, 5.0, 5.0
eps = 0.001
converged = False
x_old = np.array([x1, y1, z1])

print("Hasil iterasi:")
for i in range(1, 101):

    x1_new = (10 - 3 * y1 + z1) / 2
    y1_new = x1_new + 2 * z1 - 4
    z1_new = 14 - 3 * x1_new - 2 * y1_new

    x_new_arr = np.array([x1_new, y1_new, z1_new])
    dx = np.linalg.norm(x_new_arr - x_old)

    print("%d, %.4f, %.4f, %.4f" % (i, x1_new, y1_new, z1_new))

    if dx < eps:
        converged = True
        print("Converged")
        break

    x_old = x_new_arr.copy()
    x1, y1, z1 = x1_new, y1_new, z1_new

if not converged:
    print("Not Converged")
```

NOMOR 1 DENGAN METODE NUMPY

```python
print("\n=== System 1: Numpy Solution ===")
A1 = np.array([[2, 3, -1],
               [1, -1, 2],
               [3, 2, 1]], dtype=float)
b1 = np.array([10, 4, 14], dtype=float)

solution1 = np.linalg.solve(A1, b1)
print("Solution (x, y, z):", solution1)
```

NOMOR 2 DENGAN GAUSS SEIDELL

```python
print("\n=== System 2: Gauss-Seidel Method ===")

A2_check = np.array([[3, 0.1, 0.2],
                     [0.1, 7, 0.3],
                     [0.3, 0.2, 10]], dtype=float)

diag2 = np.diag(np.abs(A2_check))
off_diag2 = np.sum(np.abs(A2_check), axis=1) - diag2

if np.all(diag2 > off_diag2):
    print("Matrix is diagonally dominant")
else:
    print("NOT diagonally dominant")


x2, y2, z2 = 1.0, 2.0, 3.0
eps2 = 0.001
converged2 = False
x2_old = np.array([x2, y2, z2])

print("Hasil iterasi:")
for i in range(1, 101):
    x2_new = (7.85 + 0.1 * y2 + 0.2 * z2) / 3
    y2_new = (-19.3 - 0.1 * x2_new + 0.3 * z2) / 7
    z2_new = (71.4 - 0.3 * x2_new + 0.2 * y2_new) / 10

    x2_new_arr = np.array([x2_new, y2_new, z2_new])
    dx2 = np.linalg.norm(x2_new_arr - x2_old)

    print("%d, %.4f, %.4f, %.4f" % (i, x2_new, y2_new, z2_new))

    if dx2 < eps2:
        converged2 = True
        print("Converged")
        break

    x2_old = x2_new_arr.copy()
    x2, y2, z2 = x2_new, y2_new, z2_new

if not converged2:
    print("Not Converged")
```

NOMOR 2 DENGAN NUMPY

```python
print("\n=== System 2: Numpy Solution ===")
A2 = np.array([[3, -0.1, -0.2],
               [0.1, 7, -0.3],
               [0.3, -0.2, 10]], dtype=float)
b2 = np.array([7.85, -19.3, 71.4], dtype=float)

solution2 = np.linalg.solve(A2, b2)
print("Solution (x, y, z):", solution2)
```

NOMOR 3 REGRESI LINEAR DENGAN LEAST SQUEARES

```python
import matplotlib.pyplot as plt

print("\n=== Regression with Least Squares ===")

X = np.array([5, 4, 6, 7, 5], dtype=float)
Y = np.array([10, 8, 12, 14, 10], dtype=float)

#Hitung mean
mean_X = np.mean(X)
mean_Y = np.mean(Y)


a_reg = np.sum((X - mean_X) * (Y - mean_Y)) / np.sum((X - mean_X) ** 2)
b_reg = mean_Y - a_reg * mean_X

print("Nilai a (slope) = {:.4f}".format(a_reg))
print("Nilai b (intercept) = {:.4f}".format(b_reg))

# Plot data dan garis regresi
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Data Titik')
x_fit = np.linspace(min(X)-1, max(X)+1, 100)
y_fit = a_reg * x_fit + b_reg
plt.plot(x_fit, y_fit, color='red', label='Garis Regresi')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresi Linear: y = ax + b")
plt.legend()
plt.grid(True)
plt.show()
```
