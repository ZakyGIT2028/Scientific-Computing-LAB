```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda t, s: np.exp(-t)
h = 0.1
t_min = 0
t_max = 1

t = np.arange(t_min, t_max + h, h)
s0 = -1
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'r', label='Exact')
plt.title('Approximate and Exact Solution')
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta_4th_order(f, x0, y0, h, x_final):
  x_values = [x0]
  y_values = [y0]
  x = x0
  y = y0

  while x < x_final:
    k1 = h * f(x, y)
    k2 = h * f(x + 0.5 * h, y + 0.5 * k1)
    k3 = h * f(x + 0.5 * h, y + 0.5 * k2)
    k4 = h * f(x + h, y + k3)
    y = y + (1/6) * (k1 + 2 * k2 + 2 * k3 + k4)
    x = x + h
    x_values.append(x)
    y_values.append(y)

  return x_values, y_values

def f(x, y):
  return x**2 - 4*y

x0 = 0
y0 = 1
h = 0.1
x_final = 1

x_rk, y_rk = runge_kutta_4th_order(f, x0, y0, h, x_final)

plt.figure(figsize=(10, 6))
plt.plot(x_rk, y_rk, marker='o', linestyle='-', label='Runge-Kutta 4th Order')
plt.title('Solution of y\' + 4y = x^2 using Runge-Kutta 4th Order Method')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend()
plt.show()
```
