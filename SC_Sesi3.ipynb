LINEAR INTERPOLASI

```python
x = [0,1,2]
y = [1,3,2]
```

NEWTON INTERPOLATION

```python
%matplotlib inline
def divided_diff(x,y):
    '''
    function to calculate the divided
    differences table
    '''

    n= len(y)
    coef = np.zeros([n,n])
    coef[:,0] = y

    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
            (coef[i+1][j-1] - coef[i][j-1])/(x[i+j]-x[i])

    return coef

def newton_poly(coef, x_data, x):
  n = len(x_data) - 1
  p = coef[n]
  for k in range(1, n+1):
    p = coef[n-k] + (x - x_data[n-k])*p
  return p
```

```python
x = np.array([-5,-1,0,2])
y = np.array([-2,6,1,3])


a_s = divided_diff(x,y)[0, :]

x_new = np.arange(-5,2.1,.1)
y_new = newton_poly(a_s, x, x_new)

plt.figure(figsize = (10,8))
plt.plot(x,y, 'bo')
plt.plot(x_new, y_new)
plt.title("Newton Interpolation")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d


distance = np.array([0, 5, 10])
speed = np.array([10, 15, 20])


interpolation = interp1d(distance, speed, kind='linear')


speed_7_2 = interpolation(7.2)
print(f'Kecepatan pada jarak 7.2 meter: {speed_7_2:.2f} m/detik')


x_new = np.linspace(0, 10, 100)
y_new = interpolation(x_new)

plt.plot(distance, speed, 'ro', label='Data')
plt.plot(x_new, y_new, 'b-', label='Interpolasi Linear')
plt.scatter(7.2, speed_7_2, color='g', label=f'7.2 m: {speed_7_2:.2f} m/s')

plt.xlabel('Jarak (m)')
plt.ylabel('Kecepatan (m/s)')
plt.title('Interpolasi Linear Kecepatan Lari')
plt.legend()
plt.grid()
plt.show()
```
