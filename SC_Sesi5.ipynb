```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-poster")
%matplotlib inline

h = 0.1
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)

forward_diff = np.diff(y) / h
x_diff = x[:-1]

exact_solution = -np.sin(x_diff)

plt.figure(figsize=(10, 6))
plt.plot(x_diff, forward_diff, label="Forward Difference")
plt.plot(x_diff, exact_solution, label="Exact Solution")
plt.legend()
plt.show()


max_error = np.max(np.abs(forward_diff - exact_solution))
print("Maximum Error:", max_error)
```

```python
import numpy as np

def f(x):
    return np.sin(x)

def derivatif(x, h=1e-5):
    return (f(x + h) - f(x)) / h

x_values = [0, 0.2]
for x in x_values:

    print(f"f'({x}) = {derivatif(x):.4f}")
```

```python
import numpy as np

def f(x):
    return np.sin(x)

def trapezium_composite(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    integral = h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1])
    return integral


a = 0
b = np.pi

# Panel 8
n1 = 8
hasil1 = trapezium_composite(a, b, n1)
print(f"Hasil dengan 8 panel: {hasil1:.6f}")

# Panel 16
n2 = 16
hasil2 = trapezium_composite(a, b, n2)
print(f"Hasil dengan 16 panel: {hasil2:.6f}")


print(f"Hasil eksak: {2.0}")
```
